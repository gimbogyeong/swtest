###1470. Shuffle the Array

Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

Return the array in the form [x1,y1,x2,y2,...,xn,yn].

Input: nums = [2,5,1,3,4,7], n = 3
Output: [2,3,5,4,1,7] 
Explanation: Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

https://leetcode.com/problems/shuffle-the-array/description/

```python
%%writefile 1470.cpp
class Solution {
public:
    vector<int> shuffle(vector<int>& nums, int n) {
        vector <int> ans; 

        for (int i = 0 ; i<n ; i++){
            ans.push_back(nums[i]);
            ans.push_back(nums[i+n]);
        }

        return ans; 
    }
};

//  뭐지? 왜 이럼? 
int main()
{
  vector<int> nums = {2,5,1,3,4,7};
  int n = 4; 

  Solution solve;
  vector<int> ret = solve.shuffle(nums ,n);

  for (auto val : ret ) {
      cout << val << ' ';
  }
  cout << endl;
}

// 헐 .... 
```

###1480. Running Sum of 1d Array

Given an array nums. We define a running sum of an array as runningSum[i] = sum(nums[0]…nums[i]).

Return the running sum of nums.


[문제바로가기](https://leetcode.com/problems/running-sum-of-1d-array/)

```python
class Solution {
public:
    vector<int> runningSum(vector<int>& nums) {
        vector<int> ans; 
        for (int j= 0 ; j<nums.size(); j++){
            int sum = 0;

            for (int i =0 ; i<=j; i++){
                  sum = sum+nums[i]; 
            }   
            ans.push_back(sum); 
        }
        return ans; 
    }
};
```

```python
vector<int> nums = {1,2,3,4};

Solution solve;
vector<int> ret = solve.runningSum( nums );

for (auto val : ret ) {
    cout << val << ' ';
}
cout << endl;
```

###1512. Number of Good Pairs

[문제바로가기](https://leetcode.com/problems/number-of-good-pairs/submissions/849533572/)

```python
class Solution {
public:
    int numIdenticalPairs(vector<int>& nums) {

        int good_pairs = 0; 
        for (int i =0 ; i<nums.size(); i++){
            for (int j =i+1 ; j <nums.size() ; j++){
                if (nums[i] == nums[j]){
                    good_pairs++; 
                }
            }
        }
        return good_pairs; 
    }
};
```

```python
vector<int> nums = {[[1,2,3,1,1,3]};

Solution solve;
vector<int> ret = solve.numIdenticalPairs( nums );

for (auto val : ret ) {
    cout << val << ' ';
}
cout << endl;
```


###1929. Concatenation of Array

[문제 바로가기](https://leetcode.com/problems/concatenation-of-array/submissions/850992981/)

```python
class Solution {
public:
    vector<int> getConcatenation(vector<int>& nums) {
        vector<int> ans; 

        for(int i = 0 ; i<nums.size() ; i++){
            ans.push_back(nums[i]);
        }

        for(int i = 0 ; i<nums.size() ; i++){
            ans.push_back(nums[i]);
        }
        return ans; 
    }
};
```

```python
vector<int> nums = {[1,2,1]};

Solution solve;
vector<int> ret = solve.getConcatenation( nums );

for (auto val : ret ) {
    cout << val << ' ';
}
cout << endl;
```

###1920. Build Array from Permutation

[문제 바로가기](https://leetcode.com/problems/build-array-from-permutation/submissions/850988080/)


```python
class Solution {
public:
    vector<int> buildArray(vector<int>& nums) {
        vector<int> ans;

        for (int i = 0 ; i< nums.size() ; i++){
            ans.push_back(nums[nums[i]]);
        }
    return ans; 
    }
        
};
```

```python
vector<int> nums = {[0,2,1,5,3,4]};

Solution solve;
vector<int> ret = solve.buildArray( nums );

for (auto val : ret ) {
    cout << val << ' ';
}
cout << endl;
```

###1672. Richest Customer Wealth

Input: accounts = [[1,2,3],[3,2,1]]

Output: 6

Explanation:

1st customer has wealth = 1 + 2 + 3 = 6

2nd customer has wealth = 3 + 2 + 1 = 6

[문제 바로가기](https://leetcode.com/problems/richest-customer-wealth/description/)

```python

%%writefile b.cpp 


#include <iostream>
#include <vector> 

using namespace std; 

class Solution {
public:
    int maximumWealth(vector<vector<int>>& accounts) {

        vector <int>  sum = {0};
         
        for (int i = 0 ; i < accounts.size(); i++){
            int sum_cal = 0 ; 
            for (int j = 0 ; j<accounts[i].size(); j++){
                sum_cal += accounts[i][j] ;
            }
            sum.push_back( sum_cal ); 
        }
        
        int max = -100; 
        for (int i = 0 ; i < sum.size();  i++){
            if(max <sum[i]){
                max = sum[i]; 
            }
        }
        return max; 
    }
};

int main()
{
    // printf("Hello World");
    
    // int accounts[2][2] ;

    vector<vector<int>> accounts ={{1,2,3},{3,2,1}};
    Solution solve; 
    

    int ret = solve.maximumWealth(accounts); 
    
    cout << ret << endl; //endl은 줄바꿈  

    return 0;
}
```
